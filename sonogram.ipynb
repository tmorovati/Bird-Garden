## libraries

```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy import signal
import numpy as np
import os 
```

## read song file

```python
# Read the voice file
address = "C:/My_Document/University_Courses/papers/IPM-started@Tir25th/python-code/NickleDave/Bird0/Wave/"
samplingFrequency, signalData = wavfile.read( address + '0.wav')
```

## Compute the spectrogram

```python

frequencies, times, spectrogram = signal.spectrogram(signalData, samplingFrequency)

# Plot the spectrogram as a sonogram
plt.pcolormesh(times, frequencies, np.log10(spectrogram))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

## read a list of wav file

```python
# Get a list of all the wav files in the folder
wav_files = [f for f in os.listdir(address) if f.endswith('.wav')]

# Loop over all the wav files and plot their spectrograms
for wav_file in wav_files:
    # Read the wav file
    file_path = os.path.join(address, wav_file)
    samplingFrequency, signalData = wavfile.read(file_path)

    # Compute the spectrogram
    frequencies, times, spectrogram = signal.spectrogram(signalData, samplingFrequency)

    # Plot the spectrogram as a sonogram
    plt.pcolormesh(times, frequencies, np.log10(spectrogram))
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
    plt.title(wav_file)
    plt.show()
```
